```python
import psycopg2
import pandas as pd
hostname = '127.0.0.2'
username = 'emg'
database = 'postgres'
```

```python
# Connect to the database
conn = psycopg2.connect(host=hostname,user=username,dbname=database)

# Create a cursor object
cur = conn.cursor()

# Execute query
# cur.execute("SELECT * FROM matches LIMIT 5;")
cur.execute(open("match_outcome.sql", "r").read())

# Fetch the results
results = cur.fetchall()
print(results)

# Close the connection
cur.close()
conn.close()
```
